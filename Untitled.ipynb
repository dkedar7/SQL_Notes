```python
db = sqlite3.connect('chinook/chinook.db')
c=db.cursor()
```

```python
def show_as_pandas_df(query,db=db):
    return pd.read_sql_query(query,db)
```

```python
query = "SELECT name FROM sqlite_master WHERE type='table';"
show_as_pandas_df(query)
```
